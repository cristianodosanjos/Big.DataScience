```python
db.columns = ['date', 'instrument', 'open', 'close', 'low', 'high', 'volume']
db.head()
```

```python
print('The operaions in this data began on:','',db['date'].min(), 'and ended in:','',db['date'].max())
```

```python
db['difference'] = db['close'] - db['open']
db.head()
```

```python
db['downfall'] = db['close'] - db['high']
db.head()
```

The column difference show us the profit and loss in each time with every instrument, the column downfall is the drop in money.

```python
db.tail()
```

```python
dtt = db.groupby('instrument')['difference'].sum()
print(dtt.head(), dtt.tail())
```

```python
print('max is:', dtt.max(), '  and min is :', dtt.min())
```

```python
dtt = pd.DataFrame(dtt)
dt = dtt.reset_index(drop = False)
```

```python
dt.head()
```

```python
dt.query('difference > 300')
```

```python
dt.query('difference < -300')
```
